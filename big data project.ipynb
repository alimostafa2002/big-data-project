```python
import requests
from bs4 import BeautifulSoup
from concurrent.futures import ThreadPoolExecutor
import time
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python


# this part is for scraping in parallel way using beautifulsoup and threadpoolexecuter


urls = [
    "https://www.oak.com/blog/employee-engagement-statistics/"
    

    
]

def scrape_title_parallel(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    return soup.find_all('h3')

start_time = time.time()
with ThreadPoolExecutor(max_workers=100000000) as executor:
    titles = list(executor.map(scrape_title_parallel, urls))
for title in titles:
    print(f"div {title}")
end_time = time.time()

```
